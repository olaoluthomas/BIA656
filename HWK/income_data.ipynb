```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# %matplotlib inline
import requests
```

```python
link = "http://personal.stevens.edu/~gcreamer/BIA656/income1.data"
r = requests.get(link)
t = r.text
```

```python
test = t[:677]
test
```

```python
lst = []
lst_in_lst = []
for char in t:
    if len(lst) == 8993:
        break
    elif char == u'\n':
        lst.append(lst_in_lst)
        lst_in_lst = []
    elif char == u' ':
        continue
    elif char == u'N':
        lst_in_lst.append(np.nan)
    elif char == u'A':
        continue
    else:
        lst_in_lst.append(char)
```

```python
print lst_in_lst
print lst
```

```python
for l in lst:
    if len(l) != 14:
        print lst.index(l)
```
